```python
# importing the packages
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

# set the style of the plots to whitegrid
sns.set_style(style="whitegrid")
```

```python
# read the data 
data = pd.read_csv("../data/Churn_Modelling.csv")

# drop the rownumber, custommerid, and name columns]
data.drop(columns=['RowNumber', 'CustomerId', 'name'], inplace=True)

# display the first five records of the dataframe
data.head()
```

```python
plt.figure(figsize=(10,3))
sns.boxplot(x=data["Age"], data=data, color='gray', linewidth=1)
sns.despine(left=True)
plt.title("Age Box Plot")
plt.savefig("charts/boxplot_age.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(x=data['Age'], bins=30, color='gray', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.savefig("charts/histogram_age.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.countplot(x=data["Tenure"], data=data, color='gray', linewidth=1)
sns.despine(left=True)
plt.title("Tenure Count Plot")
plt.savefig("charts/countplot_tenure.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(x=data['Balance'], bins=40, color='gray', edgecolor='black')
plt.xlabel("Account Balance")
plt.ylabel("Frequency")
plt.title("Histogram of Balance in Account")
plt.savefig("charts/histogram_account_balance.jpg")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.countplot(x=data["NumOfProducts"], data=data, color='gray', linewidth=1)
sns.despine(left=True)
plt.title("Product Subscription Count Plot")
plt.savefig("charts/countplot_products_subscribed.jpg")
plt.show()
```

```python
plt.figure(figsize=(4,5))
sns.countplot(x=data["HasCrCard"], data=data, color='gray', linewidth=1)
sns.despine(left=True)
plt.title("Credit Card Subscription Count Plot")
plt.savefig("charts/countplot_credit_card_subscribed.jpg")
plt.show()
```

```python
plt.figure(figsize=(4,5))
sns.countplot(x=data["IsActiveMember"], data=data, color='gray', linewidth=1)
sns.despine(left=True)
plt.title("Is Active Member Count Plot")
plt.savefig("charts/countplot_is_active_member.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(x=data['EstimatedSalary'], bins=100, color='gray', edgecolor='black')
plt.xlabel("Estimated Salary")
plt.ylabel("Frequency")
plt.title("Histogram of Estimated Salary")
plt.savefig("charts/histogram_estimated_salary.jpg")
plt.show()
```
